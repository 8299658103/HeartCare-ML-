```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# 2. Load Dataset
df = pd.read_csv('Heart_disease_details.csv')

# 3. Overview
print("ðŸ”¹ Shape of data:", df.shape)
print("\nðŸ”¹ Columns:\n", df.columns.tolist())
print("\nðŸ”¹ Missing values:\n", df.isnull().sum())
print("\nðŸ”¹ Data types:\n", df.dtypes)
print("\nðŸ”¹ Sample Data:\n", df.head())
```

```python
# DATA CLEANING
df = df.drop_duplicates()

# Handle categorical column standardization (strip spaces, lower case)
df.columns = df.columns.str.strip()

# Example: Convert gender to binary
df["Gender"] = df["Gender"].replace({"Male": 1, "Female": 0}).astype(int)

# Fill missing values (example strategies)
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Cholesterol level (mg/dL)"] = df["Cholesterol level (mg/dL)"].fillna(df["Cholesterol level (mg/dL)"].mean())
```

```python
# UNIVARIATE ANALYSIS

# Age distribution
sns.histplot(df["Age"], bins=30, kde=True, color="blue")
plt.title("Age Distribution")
plt.show()

# Gender distribution
sns.countplot(x="Gender", data=df,hue="Gender", palette="pastel", legend=False)
plt.title("Gender Count")
plt.xticks([0, 1], ["Female", "Male"])
plt.show()

# Diagnosis distribution
sns.countplot(x="Diagnosis", data=df, hue="Diagnosis", palette="coolwarm", legend=False)
plt.title("Diagnosis Distribution")
plt.xticks(rotation=90)
plt.show()
```

```python
# ðŸ“Š BIVARIATE ANALYSIS

# Age vs Diagnosis
sns.boxplot(x="Diagnosis", y="Age", data=df, palette="Set2")
plt.xticks(rotation=90)
plt.title("Age vs Diagnosis")
plt.show()

# Gender vs Diagnosis
sns.countplot(x="Diagnosis", hue="Gender", data=df, palette="Set1")
plt.xticks(rotation=90)
plt.title("Gender vs Diagnosis")
plt.show()

# Cholesterol vs Diagnosis
sns.boxplot(x="Diagnosis", y="Cholesterol level (mg/dL)", data=df, palette="Set3")
plt.xticks(rotation=90)
plt.title("Cholesterol vs Diagnosis")
plt.show()
```

```python
# CORRELATION ANALYSIS
# Select only numeric columns
num_df = df.select_dtypes(include=np.number)

# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(num_df.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
# FEATURE INSIGHTS

# Hypertension vs Diagnosis
sns.countplot(x="Hypertension", hue="Diagnosis", data=df, palette="muted")
plt.xticks([0, 1], ["No Hypertension", "Hypertension"])
plt.title("Hypertension vs Diagnosis")
plt.show()
```

```python
# Smoking vs Diagnosis
sns.countplot(x="Smoking", hue="Diagnosis", data=df, palette="husl")
plt.xticks([0, 1], ["Non-Smoker", "Smoker"])
plt.title("Smoking vs Diagnosis")
plt.show()
```

```python
# Diabetes vs Diagnosis
sns.countplot(x="Diabetes", hue="Diagnosis", data=df, palette="viridis")
plt.xticks([0, 1], ["Non-Diabetic", "Diabetic"])
plt.title("Diabetes vs Diagnosis")
plt.show()
```
